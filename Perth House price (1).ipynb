```python
perth= train_set.drop('PRICE', axis=1)
training_labels= train_set['PRICE'].copy()
```

```python
perth.shape, training_labels.shape
```

```python
pd.isnull(perth).sum()
```

```python
##Handling missing values and Normalizing the data.
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
numerical_pipeline= Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler())
])
```

```python
from sklearn.compose import ColumnTransformer
perth.columns
```

```python
num_features=('BEDROOMS', 'BATHROOMS', 'GARAGE', 'LAND_AREA', 'FLOOR_AREA','BUILD_YEAR', 'CBD_DIST', 'NEAREST_SCH_RANK')
```

```python
preprocessings= ColumnTransformer([
    ('num', numerical_pipeline, num_features)
])
```

```python
final_data= preprocessings.fit_transform(perth)
final_data.shape
```

```python
preprocessings.get_feature_names_out()
```

```python
##Model Selection: Where first model is linear regression.
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
```

```python
lin_reg= make_pipeline(preprocessings, LinearRegression())
lin_reg.fit(perth, training_labels)
```

```python
predictions_reg= lin_reg.predict(perth)
predictions_reg[:10]
```

```python
training_labels[:10].values
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score_reg= r2_score(training_labels, predictions_reg)
```

```python
r2_score_reg
```

```python
## USing Random Forest Regressor for Predictions
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
forrest_reg= make_pipeline(preprocessings, RandomForestRegressor(random_state=42))
r2_score= cross_val_score(forrest_reg, perth, training_labels, scoring='r2', cv=10)
```
