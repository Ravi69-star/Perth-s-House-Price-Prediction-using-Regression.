##Data Preprocessing and feature engineering using Pipelines.

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
numerical_pipeline= Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler())
])
```

```python
from sklearn.compose import ColumnTransformer
perth.columns
```

```python
num_features=('BEDROOMS', 'BATHROOMS', 'GARAGE', 'LAND_AREA', 'FLOOR_AREA','BUILD_YEAR', 'CBD_DIST', 'NEAREST_SCH_RANK')
```

```python
preprocessings= ColumnTransformer([
    ('num', numerical_pipeline, num_features)
])
```

```python
final_data= preprocessings.fit_transform(perth)
final_data.shape
```

```python
preprocessings.get_feature_names_out()
```

##Fitting the data into linear regression model.

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
```

```python
lin_reg= make_pipeline(preprocessings, LinearRegression())
lin_reg.fit(X_train, y_train)
```

```python
predictions_reg= lin_reg.predict(X_test)
predictions_reg[:10]
```

```python
y_test[:10].values
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
r2_score_reg= r2_score(y_test, predictions_reg)
r2_score_reg
```

```python
lin_mse= mean_squared_error(y_test, predictions_reg)
lin_mse
```

##The r2 scoere is around 54 percentage whereas mse is also very high so, lets try some other models for non-linearity.

###Using Random Forest regressor.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
forrest_reg= make_pipeline(preprocessings, RandomForestRegressor(random_state=42))
forrest_reg.fit(X_train, y_train)
```

```python
forrest_reg_predict= forrest_reg.predict(X_test)
forrest_reg_predict[:5]
```

```python
y_test[:5].values
```

```python
r2_forest= r2_score(y_test, forrest_reg_predict)
r2_forest
```

```python
forest_mse= mean_squared_error(y_test, forrest_reg_predict)
forest_mse
```

```python
r2_c_val= cross_val_score(forrest_reg, X_train, y_train, scoring='r2', cv=5)
```

```python
print(r2_c_val)
```

##We can see this model is doing better job than linear one where r2 score is 75+ and mse is also very low and, I used cross validation to see any overfitting into the model and to wnsure the model does good during the unseen data.

##Hyperparameter tuning is optional and I am using automatic parameter selection method using GridSearchCV.

```python
from sklearn.model_selection import GridSearchCV
```

```python
reg_forest= make_pipeline(preprocessings, RandomForestRegressor(random_state=42))
```

##I have give some random parameters and we can also some more like leaf etc but I am just using these.

```python
param_grid= {
    'randomforestregressor__n_estimators': [25, 50, 75, 100, 150, 200],
    'randomforestregressor__max_depth': [None, 2, 4, 6, 8, 10],
    'randomforestregressor__max_features': [4, 5, 6, 8, 9, 10]
}
```

```python
grid_search= GridSearchCV(reg_forest, param_grid, cv=3, scoring= 'r2')
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_, grid_search.best_score_
```

```python
best_model= grid_search.best_estimator_
```

```python
best_model_prediction= best_model.predict(X_test)
best_model_prediction[:5]
```

```python
r2_tuned_model= r2_score(y_test, best_model_prediction)
r2_tuned_model
```

```python
mse_tuned_model= mean_squared_error(y_test, best_model_prediction)
mse_tuned_model
```

##Not a huge improvement but yes there is some better result in our tuned model.
